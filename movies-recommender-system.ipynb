```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
credits.head()
```

```python
credits.shape
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

```python
# important columns
# movie_id
# title
# overview
# genres
# keywords
# cast 
# crew
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'][0]
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
m = movies['cast'][0]
```

```python
convert(m)
```

```python
def fetch_3_actors(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
fetch_3_actors(m)
```

```python
movies['cast'] = movies['cast'].apply(fetch_3_actors)
```

```python
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# removing spaces - Sam Worthington to SamWorthington
def remove_space(List):
    List1 = []
    for i in List:
        List1.append(i.replace(" ",""))
    return List1
```

```python
remove_space(['Sam Worthington'])
```

```python
movies['cast'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
final_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
final_df.head()
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x: " ".join(x))
final_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(final_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
x = cv.get_feature_names_out()
```

```python
pd.set_option('display.max_rows', None)
pd.DataFrame(x)
```

```python
# here we will do stemming - ['love', 'loving', 'loved'] after applying stemming we'll get ['love']
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
stem('love loved loving')
```

```python
final_df['tags'] = final_df['tags'].apply(stem)
```

```python
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[0]
```

```python
def recommend(movie):
    index = final_df[final_df['title'] == movie].index[0]
    distances = similarity[index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(final_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man 3')
```
